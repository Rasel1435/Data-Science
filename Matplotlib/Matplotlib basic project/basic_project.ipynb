Now use x,y and z arrays to recreate the plot below. Notice the xlimits and ylimits on the inserted plot:

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

#Large
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel("Z")

# Insert
ax2.plot(x,z)
ax2.set_title('Zoom')
ax2.set_xlabel("X")
ax2.set_ylabel('Y')

ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```
