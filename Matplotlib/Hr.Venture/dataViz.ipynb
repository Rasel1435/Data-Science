```python
# import
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Data
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)
x, y
```

```python
# Line Plot
plt.figure(figsize=(20,8))
plt.plot(x, y, color='purple', linewidth=5, label="Sin Wave")
plt.plot(x, z, color='darkorange', linewidth=5, label="Cos Wave")
plt.title("Values vs Sin()")
plt.xlabel("X Values")
plt.ylabel("Sin")
plt.grid(True)
plt.ylim(-2,2)
plt.xlim(0,12)
plt.legend()
plt.show()
```

```python
## Bar PLot
lang = ['Python', 'R', 'Java', 'C++']
l = np.arange(1,5)
demands = [88, 45, 76, 66]
plt.bar(l, demands, color='darkorange')
plt.xticks(l, lang)
plt.title("Programming Languages vs Demands")
plt.show()
```

```python
# barh
lang = ['Python', 'R', 'Java', 'C++']
l = np.arange(1,5)
demands = [88, 45, 76, 66]
plt.barh(l, demands, color='darkorange')
plt.yticks(l, lang)
plt.title("Programming Languages vs Demands")
plt.show()
```

```python
# Histogram
bp = np.random.randint(80, 180, 50)
# bp[10:20] = 174
plt.hist(bp, bins=[80, 110, 130, 180])
```

```python
# Scatter Plot
x = np.random.randint(80, 180, 50)
y = np.random.randint(80, 180, 50)
plt.scatter(x, y, color='purple', marker='^', label="1st")
plt.scatter(x, y*3, color='darkorange', marker='*', label='2nd')
plt.legend()
```

```python
# Pie Chart
cpt = ["Poetry", 'Cricket', 'Football', 'Athletics']
participation = [80, 150, 130, 200]
plt.pie(participation, labels=cpt, explode=(0,0,0,0.05), autopct='%1.2f%%')
```

```python
# 3D PLot
x = np.random.randint(80, 180, 50)
y = np.random.randint(80, 180, 50)
z = np.random.randint(80, 180, 50)
fig = plt.figure(figsize=(20,8))
axes = plt.axes(projection='3d')
axes.scatter3D(x, y, z)
```

```python
x = np.arange(1,11)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure(figsize=(20,8))
axes = plt.axes(projection='3d')
axes.plot3D(x, y, z)
plt.savefig('plot3d.png')
```
